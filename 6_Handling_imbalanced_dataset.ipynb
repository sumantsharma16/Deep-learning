Imblanced dataset:

A dataset with an unequal class distribution

```python
# importing the dependencies
import pandas as pd
import numpy as np
```

```python
# loading the dataset to pandas dataframe
credit_data = pd.read_csv('/content/creditcard.csv')
```

This is highly imblanced dataset

0 ---> Legit
1 ---> fradulent

```python
# seprating the legit and fradulent transaction
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

```python
legit_sample = legit.sample(n=2)
```

```python
print(legit_sample.shape)
```

```python
# concatenate two dataset
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
