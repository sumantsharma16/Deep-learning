Label encoding

converting the label into numeric form

```python
# importing dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label encoding of breast cancer dataset

```python
# loading the data from csv file to pandas dataframe
```

```python
cancer_data = pd.read_csv('/content/breast cancer.csv')
```

```python
# first five rows of the data frame
```

```python
cancer_data.head()
```

```python
# diagnosis is label in this dataset
```

```python
# finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
# load the Label Encoder function
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the dataframe
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

0---> Benign


1---> Malignant

```python
cancer_data['target'].value_counts()
```

Label encoding of iris data

```python
# loading the dataframe in pandas dataframe
iris_data = pd.read_csv('/content/IRIS.csv')
```

```python
iris_data.head()
```

```python
iris_data['species'].value_counts()
```
